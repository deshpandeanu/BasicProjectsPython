**Task:** Below are the steps:

- Build a Number guessing game, in which the user selects a range.
- Let’s say User selected a range, i.e., from A to B, where A and B belong to Integer.
- Some random integer will be selected by the system and the user has to guess that integer in the minimum number of guesses

```python
import math
import random

# Taking inputs:


# generating the random number between the bounds given by the user. 
try:
    lower_bound = int(input('Enter the lower bound: '))

    upper_bound = int(input('Enter the upper bound: '))
    x = random.randint(lower_bound,upper_bound)
    
    
except ValueError:
    
    print("Oops! Looks like upper is smaller than the lower bound, please try again")

else: 
    
    number_of_guesses= round(math.log(upper_bound - lower_bound + 1, 2))
    
    print('You have',number_of_guesses,' chances to guess the number')


# inititalizing the number of guesses

    count=0

#for calculation of minimum number of
# guesses depends upon range      
    while count< number_of_guesses:
        guess = int(input("Guess a number:- "))
        if guess>upper_bound or guess<lower_bound:
            print('Guess a number within the range')
            continue
        count+=1
# Condition testing
        if x == guess:
            print("Congratulations you did it in ",count, " try")
        # Once guessed, loop will break
            break
        elif x > guess:
            print("You guessed too small!")
        elif x < guess:
            print("You Guessed too high!")
    
# If Guessing is more than required guesses,
# shows this output.
    if count >= round(math.log(upper_bound - lower_bound + 1, 2)):
        print("\nThe number is %d" % x)
        print("\tBetter Luck Next time!")
 
```
